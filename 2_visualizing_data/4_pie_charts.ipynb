```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
nbins = 5
total_N = 100

raw_data = np.ceil(np.logspace(np.log10(0.5), np.log10(nbins - 0.01), total_N))

unique_nums = np.unique(raw_data)
data_for_pie = np.zeros(len(unique_nums))

for i in range(len(unique_nums)):
    data_for_pie[i] = np.sum(raw_data == unique_nums[i])
```

```python
plt.pie(data_for_pie, labels=['one', 'two', 'three', 'four', 'five'], explode=[0, 0, 0, 0, 0.3], shadow=True, autopct='%.2f%%')
plt.show()
```
