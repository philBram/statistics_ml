```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
n = 1000

data = np.exp(np.random.randn(n) / 2)
plt.plot(data, 's')
```

```python
k = 40

plt.hist(data, bins=k)
plt.show()
```

```python
y, x = np.histogram(data, k)

#x = (x[:-1] + (x[1:] - x[:-1])) / 2
x = (x[:-1] + x[1:]) / 2

plt.plot(x, y)
```

```python
y = y / np.sum(y)

plt.plot(x, y)
```

```python
sns.histplot(data, stat='p', bins=k, kde=True)
plt.fill_between(x, 0, y)
```
