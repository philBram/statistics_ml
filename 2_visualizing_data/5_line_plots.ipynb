```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000

data_1 = np.exp(np.random.randn(n) / 2)
data_2 = np.exp(np.random.randn(n) / 10)
data_3 = np.exp(np.random.randn(n) / 2 + 1)
```

```python
k = 20

plt.hist(data_1, bins=k, alpha=0.6)
plt.hist(data_2, bins=k, alpha=0.6)
plt.hist(data_3, bins=k, alpha=0.6)

plt.show()
```

```python
y_1, x_1 = np.histogram(data_1, k)
x_1 = (x_1[:-1] + x_1[1:]) / 2
y_1 = y_1 / np.sum(y_1)

y_2, x_2 = np.histogram(data_2, k)
x_2 = (x_2[:-1] + x_2[1:]) / 2
y_2 = y_2 / np.sum(y_2)

y_3, x_3 = np.histogram(data_3, k)
x_3 = (x_3[:-1] + x_3[1:]) / 2
y_3 = y_3 / np.sum(y_3)

plt.plot(x_1, y_1, 's-',label='data_1')
plt.fill_between(x_1, y_1, alpha=0.4)

plt.plot(x_2, y_2, 'o-',label='data_2')
plt.fill_between(x_2, y_2, alpha=0.4)

plt.plot(x_3, y_3, '^-',label='data_3')
plt.fill_between(x_3, y_3, alpha=0.4)

plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability')

plt.show()
```

```python
n = 4
x = np.linspace(0, n)

y_lin = x
y_log = np.exp(x)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].plot(x, y_lin, label='y=x')
ax[0].plot(x, y_log, label='y=e$^x$')
ax[0].set_title('lin-scaled')

ax[0].legend()

ax[1].set_yscale('log')
ax[1].plot(x, y_lin, label='y=x')
ax[1].plot(x, y_log, label='y=e$^x$')
ax[1].set_title('log-scaled')

ax[1].legend()

plt.show()
```
