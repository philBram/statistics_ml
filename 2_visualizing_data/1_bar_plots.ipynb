```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
m = 30
n = 6

data = np.zeros((m, n))

for i in range(n):
    data[:,i] = 30 * np.random.randn(m) * (2 * i / (n - 1)) ** 2 + (i + 1) ** 2
```

```python
plt.bar(range(n), np.mean(data, axis=0))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(8, 2))

ax[0].bar(range(n), np.mean(data, axis=0))
ax[0].set_title('Bar plot')

ax[1].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[1].set_title('Error plot')

ax[2].bar(range(n), np.mean(data, axis=0))
ax[2].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='', color='orange')
ax[2].set_title('Error+bar plot')

plt.show()
```

```python
xcrossings = [1, 2, 4, 5, 6, 9]

plt.bar(xcrossings, np.mean(data, axis=0))
plt.errorbar(xcrossings, np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='', color='r')
plt.xticks(xcrossings)
plt.title('Error+bar plot')

plt.show()
```

```python
m = np.array([
    [2, 5, 4, 3],
    [1, 1, 8, 6]
])
    
fig, ax = plt.subplots(2, 2, figsize=(8, 8))

ax[0, 0].imshow(m)

df = pd.DataFrame(m, columns=['prob 0', 'prob 1', 'prob 2', 'prob 3'])

df.plot(ax=ax[1, 0], kind='bar')
ax[1, 0].set_title('Grouping by rows')

ax[0, 1].imshow(m.T)

df.T.plot(ax=ax[1, 1], kind='bar')
ax[1, 1].set_title('Grouping by colums')

plt.show()
```
