```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 100
r = 0.6

x = np.random.randn(N)
y = np.random.randn(N)

y = x * r + y * np.sqrt(1 - r ** 2)

plt.plot(x, y, 'kp', markerfacecolor='b', markersize=12)
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
emp_r = np.corrcoef(x, y)
print(f'Desired r={r}, empirical r={emp_r[0, 1]:.5f}')
```

```python
Ns = np.round(np.linspace(5, 400, 123)).astype(int)

r = 0.6

corrs = np.zeros(len(Ns))

for n_i in range(len(Ns)):
    x = np.random.randn(Ns[n_i])
    y = x * r + np.random.randn(Ns[n_i]) * np.sqrt(1 - r ** 2)
    corrs[n_i] = (r - np.corrcoef(x, y)[0, 1]) ** 2
    
plt.stem(Ns, corrs, 'ko', basefmt=' ')
plt.xlabel('Sample size')
plt.ylabel('Squared divergence')

plt.show()
```
