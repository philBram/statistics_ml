```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
average_group_1 = 49
average_group_2 = 49

standard_deviation = 5.6

samples_group_1 = 40
samples_group_2 = 35
```

```python
data_group_1 = np.random.randn(samples_group_1) * standard_deviation + average_group_1
data_group_2 = np.random.randn(samples_group_2) * standard_deviation + average_group_2

ns = np.array([samples_group_1, samples_group_2])

datalims = [np.min(np.hstack((data_group_1, data_group_2))), np.max(np.hstack((data_group_1, data_group_2)))]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(6, 4))

ax[0].violinplot(data_group_1)
ax[0].plot(1 + np.random.randn(samples_group_1) / 10, data_group_1, 'ko')
ax[0].set_ylim(datalims)
ax[0].axis('off')

ax[1].violinplot(data_group_2)
ax[1].plot(1 + np.random.randn(samples_group_2) / 10, data_group_2, 'ko')
ax[1].set_ylim(datalims)
ax[1].axis('off')

t, p = stats.ttest_ind(data_group_1, data_group_2)

sigtxt = ('', 'Not')
fig.suptitle(f'The two groups are {sigtxt[int(p > .05)]} significantly different! t({sum(ns) - 2}), p={np.round(p, 3)}')

plt.show()
```
