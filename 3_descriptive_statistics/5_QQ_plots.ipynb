```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
n = 1000
#data = np.random.randn(n)
data = np.exp(np.random.randn(n) * 0.8)

x = np.linspace(-4, 4, 10001)
theo_norm = stats.norm.pdf(x)
theo_norm = theo_norm / np.max(theo_norm)

empiric_norm, x_1 = np.histogram(data, 40)
x_1 = (x_1[:-1] + x_1[1:]) / 2
empiric_norm = empiric_norm / np.max(empiric_norm)

plt.plot(x_1, empiric_norm, label='Empirical')
plt.plot(x, theo_norm, label='Theoretical')

plt.legend()
plt.show()
```

```python
z_sort_data = np.sort(stats.zscore(data))
sort_normal = stats.norm.ppf(np.linspace(0, 1, n))

fig, ax = plt.subplots(2, 1)

ax[0].plot(z_sort_data, 'o')
ax[0].set_title('zscore from sorted data')

ax[1].plot(sort_normal, 'o')
ax[1].set_title('zscore from interval [0, 1]')

plt.tight_layout()
plt.show()
```

```python
plt.plot(sort_normal, z_sort_data, 'o')

x_l, x_r = plt.xlim()
y_l, y_r = plt.ylim()
lims = [np.min([x_l, x_r, y_l, y_r]), np.max([x_l, x_r, y_l, y_r])]
plt.xlim(lims)
plt.ylim(lims)

plt.plot(lims, lims)

plt.xlabel('Theoretical normal')
plt.ylabel('Observed data')
plt.title('QQ plot')
plt.axis('square')

plt.show()
```
