```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
n = 1000
data = np.random.randn(n) ** 2

fig, ax = plt.subplots(2, 1)

ax[0].plot(data, '.')
ax[1].hist(data, 30)

plt.show()
```

```python
data_r = stats.rankdata(data) / n

plt.plot(data, data_r, '.')
```

```python
q_1 = np.argmin((data_r - 0.25) ** 2)
q_3 = np.argmin((data_r - 0.75) ** 2)

iq_vals = data[[q_1, q_3]]

iqr_1 = iq_vals[1] - iq_vals[0]

iqr_2 = stats.iqr(data)

print(iqr_1, iqr_2)
```
