```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
N = 10001
nbins = 30

data_1 = np.random.randn(N) - 1
data_2 = 3 * np.random.randn(N) 
data_3 = np.random.randn(N) + 1

y_1, x_1 = np.histogram(data_1, nbins)
x_1 = (x_1[:-1] + x_1[1:]) / 2

y_2, x_2 = np.histogram(data_2, nbins)
x_2 = (x_2[:-1] + x_2[1:]) / 2

y_3, x_3 = np.histogram(data_3, nbins)
x_3 = (x_3[:-1] + x_3[1:]) / 2
```

```python
plt.plot(x_1, y_1, 'b')
plt.plot(np.tile(np.mean(data_1), 2), [0, np.max(y_1)], '--b',)

plt.plot(x_2, y_2, 'r')
plt.plot(np.tile(np.mean(data_2), 2), [0, np.max(y_2)], '--r',)

plt.plot(x_3, y_3, 'k')
plt.plot(np.tile(np.mean(data_3), 2), [0, np.max(y_3)], '--k',)

plt.xlabel('Data values')
plt.ylabel('Data counts')

plt.show()
```

```python
plt.hist(data_1, nbins, color='b', alpha=0.4)
plt.plot(np.tile(np.mean(data_1), 2), [0, np.max(np.histogram(data_1, nbins)[0])], '--b')

plt.hist(data_2, nbins, color='r', alpha=0.4)
plt.plot(np.tile(np.mean(data_2), 2), [0, np.max(np.histogram(data_2, nbins)[0])], '--r')

plt.hist(data_3, nbins, color='k', alpha=0.4)
plt.plot(np.tile(np.mean(data_3), 2), [0, np.max(np.histogram(data_3, nbins)[0])], '--k')

plt.show()
```

```python
data_4 = np.hstack((np.random.randn(N) - 2, np.random.randn(N) + 2))

y_4, x_4 = np.histogram(data_4, nbins)
x_4 = (x_4[:-1] + x_4[1:]) / 2

plt.plot(x_4, y_4, 'b')
plt.plot(np.repeat(np.mean(data_4), 2), [0, np.max(y_4)], '--b', label='mean')
plt.plot(np.repeat(np.median(data_4), 2), [0, np.max(y_4)], '--r', label='median')

plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.legend()

plt.show()
```

```python
shift = 0
stretch = 0.7
n = 2000
nbins = 50
repeats = 0
outliers = [50, 100, 200]

data = stretch * np.random.randn(n - (repeats * len(outliers))) + shift
data = np.exp(data)
data = np.hstack((data, np.repeat(outliers, repeats)))

y, x = np.histogram(data, nbins)
x = (x[:-1] + x[1:]) / 2

fig, ax = plt.subplots(2, 1, figsize=(10, 10))

ax[0].plot(range(n), data, '.', color='grey', label='Data', alpha=0.9)
ax[0].plot([0, n], np.tile(np.mean(data), 2), '--r', label='mean')
ax[0].plot([0, n], np.tile(np.median(data), 2), '--b', label='median')

ax[1].plot(x, y, 'k')
ax[1].plot(np.repeat(np.mean(data), 2), [0, np.max(y)], '--b', label='mean')
ax[1].plot(np.tile(np.median(data), 2), [0, np.max(y)], '--r', label='median')

ax[0].legend()
ax[1].legend()

plt.show()
```

```python
data = np.round(np.random.randn(10))

unique_data = np.unique(data)

for i in range(len(unique_data)):
    print(f'{unique_data[i]} appears {np.sum(unique_data[i] == data)} times.')
    
print(f'\nThe modal value is {stats.mode(data, keepdims=False)[0]}')
```
