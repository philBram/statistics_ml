```python
import matplotlib.pyplot as plt
import numpy as np

from scipy import stats
```

```python
N = 1001

x = np.linspace(-4, 4, N)
gaus_dist = stats.norm.pdf(x)

plt.plot(x, gaus_dist)
plt.title('Analytic Gaussian (normal) distribution')

plt.show()

print(np.sum(gaus_dist) * np.diff(x[:2]))
```

```python
stretch = 1
shift = 5
n = 1000

data = stretch * np.random.randn(n) + shift

plt.hist(data, bins=25)
plt.title('Empirical normal distribution')

plt.show()

print(np.mean(data))
print(np.std(data, ddof=1))
```

```python
N = 1000

x = np.linspace(-5, 5, N)
uniform_dist = stats.uniform.pdf(x, scale=8, loc=-4)

plt.ylim(0, 0.2)
plt.plot(x, uniform_dist)
plt.title('Analytic uniform distibution')

plt.show()
```

```python
stretch = 4
shift = 0.5
n = 10000

data = stretch * np.random.rand(n) + shift - stretch / 2

fig, ax = plt.subplots(1, 2, figsize=(15, 5))

ax[0].plot(data, '.', markersize=1)
ax[0].set_title('Uniform data values')

ax[1].hist(data)
ax[1].set_title('Uniform data histogram')

plt.show()
```

```python
N = 1001

x = np.linspace(0, 10, N)
log_norm_dist = stats.lognorm.pdf(x, s=1)

plt.plot(x, log_norm_dist)
plt.title('Analytic log-normal distribution')

plt.show()
```

```python
stretch = 0.5
shift = 5
n = 2000

data = stretch * np.random.randn(n) + shift
data = np.exp(data)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].plot(data, '.')
ax[0].set_title('Log-normal data values')

ax[1].hist(data, bins=25)
ax[1].set_title('Log-normal data histogram')

plt.show()

print(f'mean = {np.mean(np.log(data))} and std = {np.std(np.log(data))} of original data')
```

```python
n = 10
p = 0.5

x = range(n + 1)
bin_dist = stats.binom.pmf(x, n, p)

plt.bar(x, bin_dist)
plt.xticks(range(n + 2))
plt.title(f'Binomial distribution n={n} p={p}')

plt.show()

print()
```

```python
n = 10
p = 0.5
x = 10000

data = np.random.binomial(n, p, x)

plt.hist(data, bins=11)
plt.title('Empirical binomial distribution')

plt.show()
```

```python
x = np.linspace(-4, 4, 1001)
df = 200
t = stats.t.pdf(x, df)

plt.plot(x, t)
plt.xlabel('t-value')
plt.ylabel('P(t | H$_0$)')
plt.title(f't({df}) distribution')

plt.show()
```

```python
num_df = 5
den_df = 100

x = np.linspace(0, 10, 10001)

f_dist = stats.f.pdf(x, num_df, den_df)

plt.plot(x, f_dist)
plt.title(f'F({num_df}, {den_df}) distribution')
plt.xlabel('F value')

plt.show()
```
