```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
n = 1000
thresh = 5

data = np.exp(np.random.randn(n))
data[data > thresh] = thresh + np.random.randn(np.sum(data > thresh)) * 0.1

plt.plot(data, '+')

plt.show()
```

```python
plt.hist(data, 30)
plt.title('Histogram')

plt.show()
```

```python
plt.violinplot(data)
plt.title('Violin')

plt.show()
```

```python
import seaborn as sns

sns.swarmplot(data=data, orient='v')

plt.show()
```

```python
import pandas as pd

n = 1000
thresh_1 = 3
thresh_2 = 5

data_1 = np.exp(np.random.randn(n))
data_1[data_1 > thresh_1] = thresh_1 + np.random.randn(np.sum(data_1 > thresh_1)) * 0.1

data_2 = np.exp(np.random.randn(n))
data_2[data_2 > thresh_2] = thresh_2 + np.random.randn(np.sum(data_2 > thresh_2)) * 0.1

df = pd.DataFrame(np.c_[data_1, data_2], columns=['data_1', 'data_2'])
df = df.stack()
df = pd.DataFrame(df)
df.reset_index(level=1, inplace=True)
df.columns = ['data', 'values']
df['split'] = 'split distributions'

df.head()
```

```python
sns.violinplot(data=df, x='split', y='values', hue='data', split=True)
plt.xlabel('')

plt.show()
```
