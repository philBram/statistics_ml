```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000
numbers = np.ceil(8 * np.random.rand(N) ** 2)

plt.plot(numbers, 'o')

plt.show()
```

```python
unique_numbers = np.unique(numbers)
probs = np.zeros(len(unique_numbers))

for i, number in enumerate(unique_numbers):
    probs[i] = np.sum(number == numbers) / N
    
entropy_value = -np.sum(probs * np.log2(probs + np.finfo(float).eps))

plt.bar(unique_numbers, probs)
plt.title(f'Entropy = {entropy_value:{.3}f}')
plt.xlabel('Data value')
plt.ylabel('Probability')

plt.show()
```

```python
N = 1123
brownnoise = np.cumsum(np.sign(np.random.randn(N)))

fig, ax = plt.subplots(2, 1, figsize=(4, 6))

ax[0].plot(brownnoise)
ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Data value')
ax[0].set_title('Brownian noise')

ax[1].hist(brownnoise, 30)
ax[1].set_xlabel('Data value')
ax[1].set_ylabel('Probability')

plt.show()
```

```python
nbins = 50

n_per_bin, bins = np.histogram(brownnoise, nbins)
probs = n_per_bin / np.sum(n_per_bin)

entropy_value = -np.sum(probs * np.log2(probs + np.finfo(float).eps))

print(f'Entropy = {entropy_value:.{3}f}')
```

```python
N = 1000

data = np.cumsum(np.sign(np.random.randn(N)))
entropy_data = np.zeros(len(range(4, 51)))

for b in range(4, 51):
    n_per_bin, bins = np.histogram(data, b)
    probs = n_per_bin / np.sum(n_per_bin)
    
    entropy_data[b - 4] = -np.sum(probs * np.log2(probs + np.finfo(float).eps))

plt.plot(range(4, 51), entropy_data, 's-')
plt.xlabel('Bins')
plt.ylabel('Entropy')
plt.xticks(np.arange(0, 51, 5))

plt.show()
```
