```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
n = 1000

k = 40

data = np.exp(np.random.randn(n) / 2)

plt.hist(data, k)

plt.xlabel('Value')
plt.ylabel('Count')

plt.show()
```

```python
h = 2 * stats.iqr(data) * n ** (-1/3)
k = int(np.ceil((np.max(data) - np.min(data)) / h))

plt.hist(data, k)
#plt.hist(data, bins='fd')

plt.xlabel('Value')
plt.ylabel('Count')
plt.title(f'F-D "rule" using {k}')

plt.show()
```

```python
import seaborn as sns

# uses FD rule by default
sns.histplot(data, kde=True)

plt.show()
```

```python
bins = np.linspace(5, n/2, 30).astype('int')

for b in bins:
    y, x = np.histogram(data, b)
    x = (x[:-1] + x[1:]) / 2
    plt.plot(x, y, '.-')

plt.title('size of histograms with different bins')

plt.show()
```

```python
for b in bins:
    y, x = np.histogram(data, b)
    x = (x[:-1] + x[1:]) / 2
    y = y / np.max(y)
    plt.plot(x, y, '.-')
    
plt.title('shape of distribution with different bins')

plt.show()
```
