```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 40

data_1 = np.exp(-np.abs(np.random.randn(N) * 3))
data_2 = np.exp(-np.abs(np.random.randn(N) * 5))
data_mean = [np.mean(data_1), np.mean(data_2)]

distance_data = np.zeros(N)

for i in range(len(distance_data)):
    distance_data[i] = np.sqrt((data_1[i] - data_mean[0]) ** 2 + (data_2[i] - data_mean[1]) ** 2)

distance_data = (distance_data - np.mean(distance_data)) / np.std(distance_data, ddof=1)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 6))

ax[0].plot(data_1, data_2, 'ko', markerfacecolor='k')
ax[0].plot(data_mean[0], data_mean[1], 'gp', markersize=12, markeredgecolor='k')
ax[0].set_xlabel('Variable x')
ax[0].set_ylabel('Variable y')

ax[1].plot(distance_data, 'ko', markerfacecolor=[0.7, 0.5, 0.3], markersize=12)
ax[1].set_xlabel('Data index')
ax[1].set_ylabel('Z distance')

plt.tight_layout()
plt.show()
```

```python
threshold = 2.5

outlisers = np.argwhere(distance_data > threshold)

ax[0].plot(data_1[outlisers], data_2[outlisers], 'rx', markersize=20)

ax[1].plot([0, N], [threshold, threshold], 'r--')
ax[1].plot(outlisers, distance_data[outlisers], 'rx', markersize=20)

fig
```
