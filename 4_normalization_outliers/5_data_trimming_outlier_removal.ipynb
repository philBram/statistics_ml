```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 40

data = np.random.randn(N)
data[data < -2] = -data[data < -2] ** 2
data[data > 2] = data[data > 2] ** 2

data_mean_centered = data - np.mean(data)

fig, ax = plt.subplots(1, 1)

ax.plot(data_mean_centered, 'k^', markerfacecolor='y', markersize=12)
ax.set_xlabel('Data index')
ax.set_ylabel('Data value')

plt.show()
```

```python
trim_pct = 5

data_cutoff = np.percentile(np.abs(data_mean_centered), 100 - trim_pct)

data_to_cut = np.argwhere(np.abs(data_mean_centered) > data_cutoff)

ax.plot(data_to_cut, data_mean_centered[data_to_cut], 'rx', markersize=12)

fig
```

```python
k_to_remove = 3

#data_sort_idx = np.argsort(np.abs(data_mean_centered))
#data_to_cut = data_mean_centered[data_sort_idx][len(data_mean_centered) - k_to_remove:]
data_sort_idx = np.argsort(np.abs(data_mean_centered))[::-1]
data_to_cut = data_sort_idx[:k_to_remove]

ax.plot(data_to_cut, data_mean_centered[data_to_cut], 'go', markersize=15, alpha=0.5)
ax.legend(['All data', f'threshold {(100 - trim_pct):.2f}%', f'{k_to_remove}-value threshold'])

fig
```
