```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 42
data = np.log(np.random.rand(N) + np.finfo(float).eps) * 234 + 934

data_min = data.min()
data_max = np.max(data)

data_s = (data - data_min) / (data_max - data_min)

fig, ax = plt.subplots(1, 2, figsize=(8, 4))

ax[0].plot(1 + np.random.randn(N) / 20, data, 'ks')
ax[0].set_xlim([0, 2])
ax[0].set_xticks([])
ax[0].set_ylabel('Original data scale')
ax[0].set_title('Original data')

ax[1].plot(1 + np.random.randn(N) / 20, data_s, 'ks')
ax[1].set_xlim([0, 2])
ax[1].set_xticks([])
ax[1].set_ylabel('Unity-normed data scale')
ax[1].set_title('Scaled data')

plt.tight_layout()
plt.show()
```

```python
plt.plot(data, data_s, 'ks')
plt.xlabel('Original')
plt.ylabel('Scaled')
plt.title(f'Correlation r = {np.corrcoef(data, data_s)[0, 0]:.2f}')

plt.show()
```

```python
new_min = 4
new_max = 8.7

data_new_s = data_s * (new_max - new_min) + new_min

plt.plot(1 + np.random.randn(N) / 20, data_new_s, 'ks')
plt.xlim([0, 2])
plt.xticks([])

plt.show()
```

```python
original_data = (((data_new_s - new_min) * (data_max - data_min)) / (new_max - new_min)) + data_min

plt.plot(1 + np.random.randn(N) / 20, original_data, 'ks')
plt.xlim([0, 2])
plt.xticks([])

plt.show()
```

```python
plt.plot(data, original_data, 'ks')
plt.xlabel('Original data')
plt.ylabel('Original restored data')
plt.title(f'Correlation r = {np.corrcoef(data, original_data)[0, 0]}')

plt.show()
```
