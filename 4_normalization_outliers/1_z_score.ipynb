```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
data = np.random.poisson(3, 1000) ** 2

data_mean = np.mean(data)
data_std = np.std(data, ddof=1)

plt.plot(data, 's', markersize=2)
plt.plot([0, len(data)], [data_mean, data_mean], 'r--', label='mean', linewidth=3)
plt.plot(np.tile(len(data) / 2, 2), [data_mean - data_std, data_mean + data_std], 'k--', label='std', linewidth=3)
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.title(f'Mean = {data_mean:.2f}; std = {data_std:.2f}')
plt.legend()

plt.show()
```

```python
#data_z = (data - data_mean) / data_std

data_z = stats.zscore(data)

data_z_mean = data_z.mean()
data_z_std = data_z.std(ddof=1)

plt.plot(data_z, 's', markersize=2)
plt.plot([0, len(data_z)], np.repeat(data_z_mean, 2), 'r--', linewidth=3, label='mean')
plt.plot(np.tile(len(data_z) / 2, 2), [data_z_mean - data_z_std, data_z_mean + data_z_std], 'k--', linewidth=3, label='std')
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.title(f'Mean = {data_z_mean:.2f}; std = {data_z_std:.2f}')

plt.show()
```

```python
plt.plot(data, data_z, 's')
plt.xlabel('Original')
plt.ylabel('Z-transformed')
plt.title(f'Correlation r = {np.corrcoef(data, data_z)[0, 0]}')

plt.show()
```
