```python
import numpy as np
import matplotlib.pyplot as plt
import pingouin as pg
import pandas as pd
import seaborn as sns
```

```python
data = np.loadtxt('TwoWayMixedANOVA_data.csv', delimiter=',')

time_point = np.repeat(['1', '2', '3'], 45)
groups = np.tile(np.repeat(['1', '2', '3'], 15), 3)
s = np.tile([str(i) for i in range(45)], 3)

df = pd.DataFrame(data=np.array(data).ravel('F'), columns=['The Data'])
df['Group'] = groups
df['Time point'] = time_point
df['Subject'] = s

pd.set_option('display.max_rows', None, 'display.max_columns', None)
df
```

```python
pg.mixed_anova(data=df, dv='The Data', between='Group', within='Time point', subject='Subject')
```

```python
sns.boxplot(data=df, x='Time point', y='The Data', hue='Group')
```
