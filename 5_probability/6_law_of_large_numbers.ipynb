```python
population_n = 1000000
population = np.random.rand(population_n)
population = population - np.mean(population)

sample_size = 30
sample_reps = 500
sample_means = np.zeros(sample_reps)

for i in range(sample_reps):
    idx = np.random.randint(0, population_n, sample_size)
    sample_means[i] = np.mean(population[idx])

fig, ax = plt.subplots(2, 1, figsize=(8, 6))

ax[0].plot(sample_means, 's-', markerfacecolor='w')
ax[0].plot([0, sample_reps], np.repeat(np.mean(population), 2), 'r', linewidth=3)
ax[0].set_xlabel('Experiment number')
ax[0].set_ylabel('mean value')
ax[0].legend(['Sample means', 'Population mean'])

ax[1].plot(np.cumsum(sample_means) / np.arange(1, sample_reps + 1))
ax[1].plot([0, sample_reps], np.tile(np.mean(population), 2), 'r', linewidth=3)
ax[1].set_xlabel('Experiment number')
ax[1].set_ylabel('mean value')
ax[1].legend(['Means of sample means', 'Population mean'])

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

fig, ax = plt.subplots(1, 1, figsize=(8, 4))

sns.histplot(sample_means, ax=ax)
ax.plot([np.mean(population), np.mean(population)], [ax.get_ylim()[0], ax.get_ylim()[1]], 'r')
ax.set_xlabel('Sample mean value')
ax.set_ylabel('Count')
ax.legend(['Population mean', 'Sample mean distribution'])

plt.show()
```
