```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
x = np.linspace(-5, 5, 10101)
theo_norm_dist = stats.norm.pdf(x)

num_samples = 4000

#sample_data = np.zeros(num_samples)

#for i in range(num_samples):
#    sample_data[i] = np.random.randn()

sample_data = np.random.randn(num_samples)

plt.hist(sample_data, density=True, bins=40)
#sns.histplot(sample_data, stat='density')
plt.plot(x, theo_norm_dist, 'r', linewidth=3)
plt.xlabel('Data values')
plt.ylabel('Probability')

plt.tight_layout()
plt.show()
```

```python
population_n = 1000000
population = np.random.randn(population_n)
population = population - np.mean(population)

sample_size = 30

sample_idx = np.random.randint(population_n, size=sample_size)
sample_mean = np.mean(population[sample_idx])

sample_mean
```

```python
sample_sizes = np.arange(30, 1000)

sample_means = np.ones(len(sample_sizes))

for i in range(len(sample_sizes)):
    sample_idx = np.random.randint(population_n, size=sample_sizes[i])
    sample_means[i] = np.mean(population[sample_idx])

plt.figure(figsize=(10, 6))
    
plt.plot(sample_sizes, sample_means, 's-', markerfacecolor='w')
plt.plot(sample_sizes[[0, -1]], np.tile(np.mean(population), 2), 'r', linewidth=3)
plt.xticks(np.arange(0, 1001, 50))
plt.xlabel('sample size')
plt.ylabel('mean value')
plt.legend(['Sample means', 'Popluation mean'])

plt.show()
```
