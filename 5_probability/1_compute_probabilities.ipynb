```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
c = np.array([1, 2, 4, 3])

prob = c / np.sum(c)

print(prob)
```

```python
blue = 40
yellow = 30
orange = 20
total_marbs = blue + yellow + orange

jar = np.hstack((1 * np.ones(blue), 2 * np.ones(yellow), 3 * np.ones(orange)))

num_draws = 500
draw_color = np.zeros(num_draws)

for i in range(num_draws):
    draw_color[i] = jar[np.random.randint(len(jar))]

blue_prob = np.sum(draw_color == 1) / num_draws
yellow_prob = np.sum(draw_color == 2) / num_draws
orange_prob = np.sum(draw_color == 3) / num_draws
```

```python
plt.bar([1, 2, 3], [blue_prob, yellow_prob, orange_prob], label='Proportion')
plt.plot([0.5, 1.5], np.tile(blue / total_marbs, 2),  'b', linewidth=3, label='Probability')
plt.plot([1.5, 2.5], np.tile(yellow / total_marbs, 2), 'b', linewidth=3)
plt.plot([2.5, 3.5], np.tile(orange / total_marbs, 2), 'b', linewidth=3)

plt.xticks([1, 2, 3], ['blue', 'yellow', 'orange'])
plt.ylabel('Probability/Proportion')
plt.legend()

plt.show()
```

```python
n = 10
m = 11
probabilities = np.ones((n, m))

for i in range(n):
    for j in range(m):
        probabilities[i, j] = 1 / (1 + ((n + 1 - (i + 1)) / (j + 1)))

sns.heatmap(probabilities, xticklabels=np.arange(1, 12), yticklabels=np.arange(1, 11)[::-1])
plt.xlabel('m')
plt.ylabel('n')

plt.show()
```
