```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 10000
spike_dur = 10
spike_num_a = 0.01
spike_num_b = 0.05

spike_ts_a = np.zeros(N)
spike_ts_b = np.zeros(N)

spike_times_a = np.random.randint(0, N, int(N * spike_num_a))
spike_times_b = np.random.randint(0, N, int(N * spike_num_b))
spike_times_b[: len(spike_times_a)] = spike_times_a

for i in range(len(spike_times_a)):
    bnd_pre = int(np.max((0, spike_times_a[i] - spike_dur / 2)))
    bnd_pst = int(np.min((N, spike_times_a[i] + spike_dur / 2)))
    spike_ts_a[bnd_pre: bnd_pst] = 1

for i in range(len(spike_times_b)):
    bnd_pre = int(np.max((0, spike_times_b[i] - spike_dur / 2)))
    bnd_pst = int(np.min((N, spike_times_b[i] + spike_dur / 2)))
    spike_ts_b[bnd_pre: bnd_pst] = 1

plt.plot(range(N), spike_ts_a, range(N), spike_ts_b)
plt.ylim([0, 1.2])
plt.xlim([2000, 2500])
plt.legend(['a', 'b'])

plt.show()
```

```python
prob_a = np.sum(spike_ts_a == 1) / N
prob_b = np.sum(spike_ts_b == 1) / N
prob_ab = np.mean(spike_ts_a + spike_ts_b == 2)

prob_a_given_b = prob_ab / prob_b
prob_b_given_a = prob_ab / prob_a

print(f'P(A) = {prob_a}')
print(f'P(B) = {prob_b}')
print(f'P(A*B) = {prob_ab}')
print(f'P(A|B) = {prob_a_given_b}')
print(f'P(B|A) = {prob_b_given_a}')
```
