```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
x = np.linspace(0, 5, 1001)

p_1 = stats.lognorm.pdf(x, 1)
c_1 = stats.lognorm.cdf(x, 1)

p_2 = stats.lognorm.pdf(x, 0.1)
c_2 = stats.lognorm.cdf(x, 0.1)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

ax[0].plot(x, p_1 / np.sum(p_1), x, p_2 / np.sum(p_2))
ax[0].set_ylabel('probability')
ax[0].set_title('pdf(x)')

ax[1].plot(x, c_1, x, c_2)
ax[1].set_ylabel('probability')
ax[1].set_title('cdf(x)')

plt.tight_layout()
plt.show()
```

```python
cdf_1 = np.cumsum(p_1 * np.diff(x[0:2]))

plt.plot(x, c_1)
plt.plot(x, cdf_1, 'r--')

plt.show()
```

```python
p = 0.3
n = 10
x = np.arange(0, 11)

binom_pmf = stats.binom.pmf(x, n, p)
binom_cdf = stats.binom.cdf(x, n, p)

fig, ax = plt.subplots(1, 2, figsize=(8, 4))

ax[0].bar(x, binom_pmf)
ax[0].set_title('pmf')
ax[0].set_xticks(x)

ax[1].bar(x, binom_cdf)
ax[1].set_title('cdf')
ax[1].set_xticks(x)

fig.suptitle('Binomial distribution')

plt.tight_layout()
plt.show()
```

```python
from itertools import product, permutations, combinations

x = np.array([x for x in product(np.arange(1, 4), repeat=3)])
n = np.array([np.sum(s) for s in x])
p = np.array([0.1, 0.4, 0.5])

multinomial_pmf = stats.multinomial.pmf(x, n, p)

plt.bar([f'{i}' for i in x], multinomial_pmf)
plt.title('Multinomial pmf')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
