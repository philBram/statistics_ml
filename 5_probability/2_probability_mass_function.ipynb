```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
blue = 40
yellow = 30
orange = 20
total_marbs = blue + yellow + orange

jar = np.hstack((1 * np.ones(blue), 2 * np.ones(yellow), 3 * np.ones(orange)))

num_draw = 500
draw_color = np.zeros(num_draw)

for i in range(num_draw):
    draw_color[i] = jar[np.random.randint(len(jar))]

blue_prob = np.sum(draw_color == 1) / num_draw
yellow_prob = np.sum(draw_color == 2) / num_draw
orange_prob = np.sum(draw_color == 3) / num_draw
```

```python
plt.bar([1, 2, 3], [blue_prob, yellow_prob, orange_prob], label='Proportion')
plt.plot([0.5, 1.5], np.repeat(blue / total_marbs, 2), 'b', linewidth=3, label='Probability')
plt.plot([1.5, 2.5], np.repeat(yellow / total_marbs, 2), 'b', linewidth=3)
plt.plot([2.5, 3.5], np.repeat(orange / total_marbs, 2), 'b', linewidth=3)

plt.title('pmf')
plt.xticks([1, 2, 3], ['blue', 'yellow', 'orange'])
plt.ylabel('Proportion/Probability')
plt.legend()

plt.show()
```

```python
N = 10004
data_1 = np.cumsum(np.sign(np.random.randn(N)))
data_2 = np.cumsum(np.sign(np.random.randn(N)))

plt.plot(np.arange(N), data_1, linewidth=2)
plt.plot(np.arange(N), data_2, linewidth=2)

plt.show()
```

```python
nbins = 50

y_1, x_1 = np.histogram(data_1, nbins)
x_1 = (x_1[:-1] + x_1[1:]) / 2
y_1 = y_1 / np.sum(y_1)

y_2, x_2 = np.histogram(data_2, nbins)
x_2 = (x_2[:-1] + x_2[1:]) / 2
y_2 = y_2 / np.sum(y_2)

plt.plot(x_1, y_1, x_2, y_2, linewidth=3)
plt.title('pdf')
plt.xlabel('Data value')
plt.ylabel('Probability')
plt.legend(['ts1', 'ts2'])

plt.show()
```
